```python
x = [0, 2, 3, 4]
y = [1.8, 3, 4, 5]
t1 = [1.5, 4, 5, 6]
plt.plot(x, y, color="r", linestyle='--', marker='s', linewidth='3', label='x vs y')
plt.plot(x, t1, color="b", linestyle='--', marker='s', linewidth='3', label='output')
plt.xlabel('x = axies')
plt.ylabel('y = axies')
plt.title('My First Graph')
plt.legend()
plt.fill_between(x, y, color='g')
plt.fill_between(x, t1, color='black')
plt.grid(True)
plt.show()
```

```python

temperatures = np.array([36, 39, 34, 42, 40, 38, 33])
days = np.arange(1, len(temperatures) + 1)
average_temp = np.mean(temperatures)
plt.plot(days, temperatures, color="b", linestyle='--', marker='o', linewidth=2, label='Temperature (C)')
plt.xlabel('Days of the Week')
plt.ylabel('Temperature (C)')
plt.title('Temperature Over the Week')
plt.legend()

plt.fill_between(days, temperatures, color='skyblue', alpha=0.3)

plt.grid(True)

plt.show()

print("Temperatures over the week:", temperatures)
print(f"Average temperature: {average_temp:.2f} C")
```

```python
marks = np.array([
    [85, 90, 78],
    [75, 85, 80],
    [88, 92, 95],
    [79, 83, 77]
])
total_marks = np.sum(marks)
print(f"Total marks of all students in all subjects: {total_marks:.2f}")
mean_marks = np.mean(marks)
print(f"Mean (average) marks of all students: {mean_marks:.2f}")
min_marks = np.min(marks)
print(f"Minimum marks of all students: {min_marks:.2f}")
max_marks = np.max(marks)
print(f"Maximum marks of all students: {max_marks:.2f}")
```

```python
marks = np.array([
    [85, 90, 78],
    [75, 85, 80],
    [88, 92, 95],
    [79, 83, 77]
])

subjects = ['Artificial Intelligence', 'Web Development', 'Android Development']
for i, student_marks in enumerate(marks, start=1):
    plt.plot(subjects, student_marks, marker='o', label=f'Student {i}')
plt.xlabel('Books Names')
plt.ylabel('Marks')
plt.title('Marks of Students in Different Subjects')
plt.legend()
plt.show()
```
